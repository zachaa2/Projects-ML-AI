```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Importing the Dataset

We will be using the same problem as in homework 1, which is the binary classification problem about predicting patients with breast cancer.

```python
data = datasets.load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
# add target label
df['target'] = data.target
df.head()
```

# Task 1 - Implement Decision Tree Classifier

### Importing sklearn package

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
# feature data
X_data = data.data
X_columns = data.feature_names
x = pd.DataFrame(X_data, columns=X_columns)

# target data
y_data = data.target
y = pd.Series(y_data, name='target')

# split into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

# Decision Tree Object
clf = tree.DecisionTreeClassifier()

# fit model to data
clf = clf.fit(x_train, y_train)

# plot tree
plt.figure(figsize=(12,12))  # set plot size (denoted in inches)
tree.plot_tree(clf, fontsize=10)
```

```python
y_pred = clf.predict(x_test)
print(accuracy_score(y_pred, y_test))
```
