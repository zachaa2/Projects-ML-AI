```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Importing the Dataset

We will be using the same problem as in homework 1, which is the binary classification problem about predicting patients with breast cancer.

First load the dataset

```python
data = datasets.load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
# add target label
df['target'] = data.target
df.head()
```

# Task 1 - Implement Decision Tree Classifier

This time I will be using a Decision Tree Classifier to solve the binary classification problem. 

I will use the scikit learn package for the Decision Tree

[Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

### Importing sklearn package

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

### Split the data and fit the model to the data

```python
# feature data
X_data = data.data
X_columns = data.feature_names
x = pd.DataFrame(X_data, columns=X_columns)

# target data
y_data = data.target
y = pd.Series(y_data, name='target')

# split into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

# Decision Tree Object
clf = tree.DecisionTreeClassifier()

# fit model to data
clf = clf.fit(x_train, y_train)
```

### Plotting the Decision Tree

```python
# plot tree
plt.figure(figsize=(12,12))  # set plot size (denoted in inches)
tree.plot_tree(clf, fontsize=10)
```

Using the model to predict the target labels

```python
y_pred = clf.predict(x_test)
print(accuracy_score(y_pred, y_test))
```

### Changing model parameters

What happens when I change model parameters? First let's try and set the max_depth parameters of the model to some defined value. By default it is None, which means that the tree will be expanded until all leaves are pure. 

First try the Decision Tree Classifier with max depth of 3.

```python
# new Decision Tree Object
clf = tree.DecisionTreeClassifier(max_depth=3)

# fit model to data
clf = clf.fit(x_train, y_train)
```

Plotting the new tree

```python
# plot tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(clf, fontsize=10)
```

As displayed above, now the graph has three levels of depth below the root node.

Predicting target labels with max depth of 3

```python
y_pred = clf.predict(x_test)
print(accuracy_score(y_pred, y_test))
```

Clearly there is little to no difference between the defaut decision tree and this one (which is smaller). When looking at the leaves of the tree, a majority of the nodes have a gini index of 0, or close to it. 

Let's try making the decision tree even smaller and seeing how that affects the results. 

```python
# new Decision Tree Object
clf = tree.DecisionTreeClassifier(max_depth=1)

# fit model to data
clf = clf.fit(x_train, y_train)
```

Plotting the Tree

```python
# plot tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(clf, fontsize=10)
```

Now the tree is a graph with just one level of depth below the root node.

Predicting target labels with max depth of 1

```python
y_pred = clf.predict(x_test)
print(accuracy_score(y_pred, y_test))
```

There is certainly a drop in the accuracy score when decreasing the maximum depth of the tree. The reason the decrease in accuracy is not too much could be due to the relatively low number of data points, or that there is not a ton of attributes

**Measuring the Performance:** k-Fold cross validation

We can use K-Fold Cross Validation to divide the data into folds.

```python
# implement k-fold cross validation
from sklearn.model_selection import KFold 
```

First with the default max depth value

```python
# use three folds
k = 3

kfold = KFold(n_splits=k)

# redefine the bagging model
# new Decision Tree Object
clf = tree.DecisionTreeClassifier()

acc = []

for train_index, test_index in kfold.split(x):
    xtrain , xtest = x.iloc[train_index,:], x.iloc[test_index,:]
    ytrain , ytest = y[train_index] , y[test_index]
    
    clf.fit(xtrain, ytrain)
    
    pred = clf.predict(xtest)
    
    accuracy = accuracy_score(pred, ytest)
    acc.append(accuracy)

avg_acc = sum(acc) / k

print('accuracy of each fold - {}'.format(acc))
print('Avg accuracy : {}'.format(avg_acc)) 
```

Next with the max_depth of 1

```python
# use three folds
k = 3

kfold = KFold(n_splits=k)

# redefine the bagging model
# new Decision Tree Object
clf = tree.DecisionTreeClassifier(max_depth=1)

acc = []

for train_index, test_index in kfold.split(x):
    xtrain , xtest = x.iloc[train_index,:], x.iloc[test_index,:]
    ytrain , ytest = y[train_index] , y[test_index]
    
    clf.fit(xtrain, ytrain)
    
    pred = clf.predict(xtest)
    
    accuracy = accuracy_score(pred, ytest)
    acc.append(accuracy)

avg_acc = sum(acc) / k

print('accuracy of each fold - {}'.format(acc))
print('Avg accuracy : {}'.format(avg_acc))
```

# Task 2 - Bagging and Boosting

Now I will use ensemble techniques to solve the binary classification problem. These ensemble methods involve using more than one decision tree. There are multiple ensemble techniques to use. I will use the Bagging method, then the Boosting method.

To implement the Bagging and Boosting methods, I will use the sklearn.ensemble python package.

First: Implementing the **Bagging Ensemble** method

From the sklearn webpage:

A Bagging classifier is an ensemble meta-estimator that fits base classifiers each on random subsets of the original dataset and then aggregate their individual predictions (either by voting or by averaging) to form a final prediction. 

[Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.BaggingClassifier.html)

```python
from sklearn.ensemble import BaggingClassifier

bcf = BaggingClassifier(n_estimators=3).fit(x_train, y_train)
```

Since this is an ensemble method, multiple estimators (decision trees) are used. We can plot each decision tree individually.

**Estimator #1:**

```python
# plotting all three tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(bcf.estimators_[0], fontsize=10)
```

**Estimator #2:**

```python
# plotting all three tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(bcf.estimators_[1], fontsize=10)
```

**Estimator #3:**

```python
# plotting all three tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(bcf.estimators_[2], fontsize=10)
```

**Measuring the Performance:** k-Fold cross validation

We can use K-Fold Cross Validation to divide the data into folds, where each fold will be made of training partition and a validation partition. 

```python
# use three folds
k = 3

kfold = KFold(n_splits=k)

# redefine the bagging model
bcf = BaggingClassifier(n_estimators=3)

acc = []

for train_index, test_index in kfold.split(x):
    xtrain , xtest = x.iloc[train_index,:], x.iloc[test_index,:]
    ytrain , ytest = y[train_index] , y[test_index]
    
    bcf.fit(xtrain, ytrain)
    
    pred = bcf.predict(xtest)
    
    accuracy = accuracy_score(pred, ytest)
    acc.append(accuracy)

avg_acc = sum(acc) / k

print('accuracy of each fold - {}'.format(acc))
print('Avg accuracy : {}'.format(avg_acc))    
```

There seems to be a slight improvement by using the Bagging Classifier over the traditional Decision Tree Classifier.

Now let's use the other ensemble method: **Boosting Classifier**

From the sklearn webpage: 

An AdaBoost classifier is a meta-estimator that begins by fitting a classifier on the original dataset and then fits additional copies of the classifier on the same dataset but where the weights of incorrectly classified instances are adjusted such that subsequent classifiers focus more on difficult cases.

[Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html)

```python
# implementing the Boosting Classifier using sklearn
from sklearn.ensemble import AdaBoostClassifier

cf = AdaBoostClassifier(n_estimators=3, algorithm='SAMME').fit(x_train, y_train)
```

The Boosting Method is also an ensemble method, so multiple estimators are used. We can plot each decision tree individually.

**Estimator #1:**

```python
# plotting all three tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(cf.estimators_[0], fontsize=10)
```

**Estimator #2:**

```python
# plotting all three tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(cf.estimators_[1], fontsize=10)
```

**Estimator #3**

```python
# plotting all three tree
plt.figure(figsize=(12,12))  # set plot size
tree.plot_tree(cf.estimators_[2], fontsize=10)
```

**Measuring the Performance:** k-Fold cross validation

Once again, we can use K-Fold Cross Validation to divide the data into folds, where each fold will be made of training partition and a validation partition. 

```python
# use three folds
k = 3

kfold = KFold(n_splits=k)

# redefine the boosting model
cf = AdaBoostClassifier(n_estimators=3, algorithm='SAMME')

acc = []

for train_index, test_index in kfold.split(x):
    xtrain , xtest = x.iloc[train_index,:], x.iloc[test_index,:]
    ytrain , ytest = y[train_index] , y[test_index]
    
    cf.fit(xtrain, ytrain)
    
    pred = bcf.predict(xtest)
    
    accuracy = accuracy_score(pred, ytest)
    acc.append(accuracy)

avg_acc = sum(acc) / k

print('accuracy of each fold - {}'.format(acc))
print('Avg accuracy : {}'.format(avg_acc))
```

**Differences/Similarities**

Based off these results, the boosting algorithm provides an even greater level of accuracy. Both ensemble methods achieve an accuracy of over 90%, and both seem to outperform the standard Decision Tree Classifier on average.

It is interesting to note the different in performance between the three models. From worst to best, the models were Decision Tree Classifer, Bagging Classifier, and Ada Boost Classifier. However, the difference between the Decision Tree and Bagging Classifiers was much less than the difference between the Bagging and Boosting Classifiers. 

I think it is also important to note that the dataset imported from sklearn is actually relatively small, so these differences may become more or less drasting when dealing with larger or even different types of datasets. 

# Task 3 - Comparing Effectiveness

To compare the **effectiveness** of the three methods, I used K-Fold cross validation. This involves splitting the entire dataset into multiple folds, where each fold is made of training sets and a validation set. K-Fold cross validation is particularly good because it works well to protect against overfitting in a predictive model. This is particularly helpful in the case where the amount of data is limited, and models become more prone to overfitting. In cases where data is limited, the model may work well on a particular train-test-split done, but we'll have a hard time finding out if it will work on **unseen data** without trial and error. Cross validation techniques are used to protect against overfitting. 

**For example**, if you're solving a problem and the data has only 500 entries, the model may overfit to the data, and may not work as well on data not in the data set. The overfitting problem will not be revealed easily using the normal train-test-split, beacuse the dataset will only be split once. Say a 80-20 split is used, then 400 data points will be used for the train and just 100 on the testing set. It's not easy to know how effective the model is with these partitions. K-Fold Cross Validation can help to combat this problem. By creating multiple folds, we are calculating accuracies multiple times (once per fold). Each fold is made of its own training and validation set. In general, if the model good agerave performance using k-fold cross validation, we can be more confident that the model will work on new data, compared to other metrics lice accuracy score. 

**The cross validation metric is suitable for the problem at hand** because it can prove as a safer and more robust way of evaluating the performance of the models. The dataset imported for this problem is relatively small, which means the model is prone to overfitting. Cross validation will evaluate the model while protectinng against that problem.

**Other Metrics** 

Another metric used to evaluate the performance of models is the confusion matrix. The Confusion Matrix shows the breakdown of TP, FP, TN, FN.

TP - True Positive
FP - False Positive
TN - True Negative
FN - False Negative

Showing the breakdown of these metrics in a matrix can often be a useful source of information regarding the performance of the model. These metrics are specifically relevant to binary classifiaction problems, which involve Yes/No answers. In general, a model with a high accuracy will have a high number of TP and TN values, which make up the diagonal of the matrix. 

In this specific problem, the confusion matrix may not be the most useful method of evaluating effectiveness. As shown in my EDA section in Homework1, this dataset has a skewed number of target labels, so the numbers in the matrix may not be the most useful piece of information. Also, K-Fold cross validation serves as a better way to evaluate performance, especially for problems with small datasets as discussed above. 

Below is an implementation of rendering the confusion matrix, given the ground truth and predicted values. 

```python
from sklearn.metrics import ConfusionMatrixDisplay

# split into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20)

# Decision Tree Object
clf = tree.DecisionTreeClassifier().fit(x_train, y_train)

y_pred = clf.predict(x_test)

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

As shown by the above confusion matrix, the diagonal of the matrix is filled with relatively high numbers. This means the model's accuracy is relatively high. There is certainly a clear discrepancy between the False Negative and the False Positive values, but this could be due to the fact that there are more data points labeled as 1 compared to 0.

The Confusion Matrix certainly offers insightful information beyond just the accuracy score, but it still uses the typical train-test-split, which can fail to reveal problems due to overfitting as discussed above.
